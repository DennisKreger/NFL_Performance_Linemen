# Merging all data & Cleaning

## Center

```python
w4_C = cleaned_C.merge(plays_df, how="left")
w4_C.head()
```

```python
w4_C.drop(columns=[['possessionTeam','pff_passCoverage','pff_passCoverageType']], inplace=True, errors='ignore')
```

```python
w4_C = w4_C.merge(pff_df, how='left')
w4_C.head()
```

```python
w4_C = w4_C.merge(games_df, how='left')
w4_C.head()
```

```python
w4_C.drop(columns=['homeTeamAbbr', 'visitorTeamAbbr','gameDate', 'gameTimeEastern', 'season'], inplace=True, errors='ignore')
w4_C
```

```python
w4_C.to_csv('Resources/center_w4.csv', sep=',', index=False)
```
