# Merging all data & Cleaning

## Center

```python
w2_C = cleaned_C.merge(plays_df, how="left")
w2_C.head()
```

```python
w2_C.drop(columns=[['possessionTeam','pff_passCoverage','pff_passCoverageType']], inplace=True, errors='ignore')
```

```python
w2_C = w2_C.merge(pff_df, how='left')
w2_C.head()
```

```python
w2_C = w2_C.merge(games_df, how='left')
w2_C.head()
```

```python
w2_C.drop(columns=['homeTeamAbbr', 'visitorTeamAbbr','gameDate', 'gameTimeEastern', 'season'], inplace=True, errors='ignore')
w2_C
```

```python
w2_C.to_csv('Resources/center_w2.csv', sep=',', index=False)
```
