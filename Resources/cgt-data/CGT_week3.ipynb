# Merging all data & Cleaning

## Center

```python
w3_C = w3_C.merge(pff_df, how='left')
w3_C.head()
```

```python
w3_C.drop(columns=['homeTeamAbbr', 'visitorTeamAbbr','gameDate', 'gameTimeEastern', 'season'], inplace=True, errors='ignore')
w3_C
```

```python
w3_C.to_csv('Resources/center_w3.csv', sep=',', index=False)
```
